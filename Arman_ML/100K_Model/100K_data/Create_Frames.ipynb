```python
import pandas as pd
```

```python
resumes = pd.read_csv('../data/resumes.csv')
postings = pd.read_csv('../data/postings.csv')
display(resumes)
display(postings)
```

```python
# Truncate resumes to 200 and postings to 400. We will then add 50 synthetic resumes to the dataset
resumes = resumes.head(200)
postings = postings.head(400)
display(resumes)
display(postings)
```

```python
postings.to_csv('400_postings.csv', index=False)
```

```python
synth_resumes = pd.read_csv('50_syn_resumes.csv')
resumes = pd.concat([resumes, synth_resumes], ignore_index=True)
resumes.to_csv('250_resumes.csv', index=False)
display(resumes)
```
